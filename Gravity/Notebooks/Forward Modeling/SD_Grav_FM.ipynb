# Edgar Mine Microgravity Forward Modeling

```python
import numpy as np
from gpoly import gpoly, plot_model
import matplotlib.pyplot as plt
```

```python
'''
mine_shaft = np.array([
    [-1.22, 20.0],
    [ 1.22, 20.0],
    [ 1.22, 22.44],
    [-1.22, 22.44]
])
'''

mine_shaft = np.array([
    [-1.5, 20.0],
    [ 1.5, 20.0],
    [ 1.5, 23],
    [-1.5, 23]
])
```

```python
rho_rock = 2.6 # gneiss is 2.6-3 g/cm^3
rho_shaft = 0.0
d_rho = rho_rock - rho_shaft
```

```python
xmin, xmax = -50, 50
dx = 2

obs_x = np.linspace(xmin, xmax, (xmax-xmin)//dx + 1)
obs_z = np.zeros_like(obs_x)
obs_cave = np.c_[obs_x, obs_z]
```

```python
g_cave_v = gpoly(obs_cave, mine_shaft, density=d_rho)
```

```python
ax1, ax2 = plot_model(g_cave_v, obs_cave, mine_shaft)
#ax1.axhline(y=0.01, color='r', linestyle='--')
ax1.axhline(y=0.005, color='r', linestyle='--')
plt.tight_layout()
```
